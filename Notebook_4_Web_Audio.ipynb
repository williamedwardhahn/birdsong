## Step 1: Setup

Import the libraries we need for downloading, processing, and visualizing audio.

```python
import requests
import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display
from IPython.display import Audio, display

print("Libraries loaded!")
```

## Step 2: Helper Functions

These functions handle the repetitive tasks:
- `download_audio` — fetches a file from a URL and saves it locally
- `plot_waveform` — shows amplitude over time
- `plot_spectrogram` — shows frequency content over time

```python
def download_audio(url, file_name):
    """Download an audio file from a URL and save it locally."""
    print(f"Downloading {file_name}...")
    response = requests.get(url)
    if response.status_code == 200:
        with open(file_name, 'wb') as f:
            f.write(response.content)
        print(f"  Saved! ({len(response.content) / 1024:.0f} KB)")
    else:
        print(f"  Download failed (HTTP {response.status_code})")


def plot_waveform(y, sr, title="Waveform"):
    """Plot the waveform of an audio signal."""
    plt.figure(figsize=(10, 3))
    librosa.display.waveshow(y, sr=sr)
    plt.title(title)
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
    plt.show()


def plot_spectrogram(y, sr, title="Spectrogram"):
    """Plot the spectrogram of an audio signal."""
    S = np.abs(librosa.stft(y))
    S_db = librosa.amplitude_to_db(S, ref=np.max)
    plt.figure(figsize=(10, 4))
    librosa.display.specshow(S_db, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title(title)
    plt.show()


print("Helper functions ready.")
```

## Step 3: Download & Explore Sample Audio

We'll download three birdsong recordings from the web:
1. **Birdsong in woodland** — multiple birds in a natural setting
2. **Canary trill** — a single bird's melodic call
3. **Blackbird song** — a European blackbird singing

For each one, we'll see the waveform, spectrogram, and play the audio.

```python
# Define the audio samples to download
audio_samples = [
    {
        'url': 'https://soundcamp.org/sounds/382/birdsong_in_woodland_DGc.wav',
        'file_name': 'birdsong_in_woodland.wav',
        'description': 'Birdsong in Woodland'
    },
    {
        'url': 'https://soundcamp.org/sounds/383/canary_trill_yellow_2w6__oGr.wav',
        'file_name': 'canary.wav',
        'description': 'Canary Trill'
    },
    {
        'url': 'https://soundcamp.org/sounds/380/blackbird_song_5_g6O.wav',
        'file_name': 'blackbird.wav',
        'description': 'Blackbird Song'
    }
]

# Download, visualize, and play each sample
for sample in audio_samples:
    print(f"\n{'='*60}")
    print(f"{sample['description']}")
    print(f"{'='*60}")

    try:
        # Download the file
        download_audio(sample['url'], sample['file_name'])

        # Load with librosa (sr=None keeps the original sample rate)
        y, sr = librosa.load(sample['file_name'], sr=None)
        print(f"  Sample rate: {sr} Hz  |  Duration: {len(y)/sr:.2f}s")

        # Visualize
        plot_waveform(y, sr, title=f"{sample['description']} \u2014 Waveform")
        plot_spectrogram(y, sr, title=f"{sample['description']} \u2014 Spectrogram")

        # Play
        print("  Click play to listen:")
        display(Audio(y, rate=sr))

    except Exception as e:
        print(f"  Error: {e}")
```

## Step 4: Compare the Spectrograms

Let's put all three spectrograms side by side to see how different bird sounds look.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 4))

for i, sample in enumerate(audio_samples):
    try:
        y, sr = librosa.load(sample['file_name'], sr=None)
        S = np.abs(librosa.stft(y))
        S_db = librosa.amplitude_to_db(S, ref=np.max)
        librosa.display.specshow(S_db, sr=sr, ax=axes[i], x_axis='time', y_axis='log')
        axes[i].set_title(sample['description'], fontsize=11)
    except Exception as e:
        axes[i].set_title(f"Error: {e}")

plt.suptitle("Spectrogram Comparison", fontsize=14)
plt.tight_layout()
plt.show()

print("Notice how each bird produces different frequency patterns!")
```
