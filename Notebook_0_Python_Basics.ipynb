## 2. Field Notes — Storing Information in Variables

A naturalist labels their observations. In Python, we use **variables** — names that store values so we can refer to them later. Think of sticking a label on a specimen jar.

```python
# Record this morning's observations
species = "Northern Cardinal"
songs_recorded = 9
observation_hours = 2.5

# Calculate songs per hour
songs_per_hour = songs_recorded / observation_hours

# Display the result
print(songs_per_hour)
```

The `print()` command displays a value on screen. You can include text alongside your numbers using **f-strings** — they start with `f` and let you drop variables inside `{curly braces}`:

```python
print(f"Species: {species}")
print(f"Songs recorded: {songs_recorded}")
print(f"Songs per hour: {songs_per_hour}")
```

## 3. Species Log — Working with Text

Text in Python is called a **string**. It goes inside quotes. You can do useful things with strings — measure their length, change their case, or combine them.

```python
common_name = "Song Sparrow"
latin_name = "Melospiza melodia"

# Combine text
full_entry = common_name + " (" + latin_name + ")"
print(full_entry)

# How long is the name?
print(f"Common name has {len(common_name)} characters.")

# Convert to uppercase (like a field notebook heading)
print(common_name.upper())
```

## 4. Species Checklist — Lists

A **list** holds multiple items in order — like a checklist of species you've spotted. Lists use square brackets `[]` with items separated by commas.

```python
# Today's species checklist
checklist = ["Bachman's Sparrow", "Northern Cardinal", "Song Sparrow",
             "Swamp Sparrow", "Zebra Finch"]

print(f"Species observed today: {len(checklist)}")
print(f"First species spotted: {checklist[0]}")
print(f"Last species spotted:  {checklist[-1]}")
```

```python
# A new bird appears! Add it to the checklist.
checklist.append("Eastern Bluebird")
print(f"Updated checklist ({len(checklist)} species):")
print(checklist)
```

You can also store numbers in lists — like how many songs you recorded from each species:

```python
song_counts = [19, 9, 15, 10, 45]

print(f"Total songs recorded: {sum(song_counts)}")
print(f"Most songs from one species: {max(song_counts)}")
print(f"Fewest songs: {min(song_counts)}")
```

## 5. Walking the Trail — Loops

A **loop** repeats an action for each item in a list — like walking a trail and noting every bird you hear. The `for` keyword does this:

```python
# Print a formatted field log entry for each species
species_list = ["Bachman's Sparrow", "Northern Cardinal", "Song Sparrow",
                "Swamp Sparrow", "Zebra Finch"]
song_counts  = [19, 9, 15, 10, 45]

print("=" * 40)
print("  FIELD LOG — Song Counts")
print("=" * 40)

for i in range(len(species_list)):
    name = species_list[i]
    count = song_counts[i]
    print(f"  {name:.<25} {count} songs")

print("=" * 40)
print(f"  Total: {sum(song_counts)} songs")
```

## 6. The Naturalist's Toolkit — Dictionaries

A **dictionary** pairs labels with values — like a field guide entry. It uses curly braces `{}` with `key: value` pairs.

```python
# A field guide entry for one species
cardinal = {
    "common_name": "Northern Cardinal",
    "latin_name": "Cardinalis cardinalis",
    "songs_recorded": 9,
    "habitat": "woodland edges, gardens",
    "color": "red"
}

# Access a specific piece of information
print(f"Species: {cardinal['common_name']}")
print(f"Latin:   {cardinal['latin_name']}")
print(f"Habitat: {cardinal['habitat']}")
print(f"Songs:   {cardinal['songs_recorded']}")
```

## 7. Reusable Tools — Functions

A **function** is a reusable recipe. You define it once with `def`, then call it whenever you need it — like having a standard form you fill out for each bird.

```python
def field_entry(species, songs, hours):
    """Print a formatted field observation entry."""
    rate = songs / hours
    print(f"--- {species} ---")
    print(f"    Songs: {songs}  |  Hours: {hours}  |  Rate: {rate:.1f} songs/hr")
    print()

# Use the function for each species
field_entry("Bachman's Sparrow", 19, 3.0)
field_entry("Northern Cardinal", 9, 2.5)
field_entry("Song Sparrow", 15, 2.0)
field_entry("Swamp Sparrow", 10, 1.5)
field_entry("Zebra Finch", 45, 4.0)
```

## 8. Drawing What You See — Making Charts

Naturalists sketch what they observe. In Python, we use the `matplotlib` library to draw charts. We also use `numpy` for working with numbers efficiently.

We **import** these tools — think of it as grabbing a specific field guide off the shelf.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Bar chart: songs recorded per species
species_list = ["Bachman's\nSparrow", "Northern\nCardinal", "Song\nSparrow",
                "Swamp\nSparrow", "Zebra\nFinch"]
song_counts  = [19, 9, 15, 10, 45]

plt.figure(figsize=(8, 4))
plt.bar(species_list, song_counts, color=['#2ecc71', '#e74c3c', '#3498db', '#9b59b6', '#f39c12'])
plt.title("Songs Recorded Per Species")
plt.ylabel("Number of Songs")
plt.show()
```

### What Does a Sound Wave Look Like?

Bird songs are vibrations in the air. When we record them, we capture how the air pressure changes over time. Here's a simple wave — in the later notebooks, you'll see real birdsong waveforms that look much more complex.

```python
# A simple tone (like a single pure whistle)
time = np.linspace(0, 0.01, 500)  # 10 milliseconds of time
frequency = 440  # 440 Hz = the note A
wave = np.sin(2 * np.pi * frequency * time)

plt.figure(figsize=(10, 3))
plt.plot(time * 1000, wave, color='#2ecc71')
plt.title("A Simple Sound Wave (440 Hz — the note A)")
plt.xlabel("Time (milliseconds)")
plt.ylabel("Air Pressure")
plt.grid(True, alpha=0.3)
plt.show()

print("Real birdsong is a complex mix of many frequencies at once!")
```

```python
# Combine two tones — like a bird singing two notes at once
wave_low = np.sin(2 * np.pi * 300 * time)   # 300 Hz (low note)
wave_high = np.sin(2 * np.pi * 800 * time)  # 800 Hz (high note)
combined = wave_low + wave_high

fig, axes = plt.subplots(3, 1, figsize=(10, 6), sharex=True)

axes[0].plot(time * 1000, wave_low, color='#3498db')
axes[0].set_title("Low Note (300 Hz)")
axes[0].set_ylabel("Amplitude")

axes[1].plot(time * 1000, wave_high, color='#e74c3c')
axes[1].set_title("High Note (800 Hz)")
axes[1].set_ylabel("Amplitude")

axes[2].plot(time * 1000, combined, color='#9b59b6')
axes[2].set_title("Both Notes Combined")
axes[2].set_ylabel("Amplitude")
axes[2].set_xlabel("Time (milliseconds)")

plt.tight_layout()
plt.show()

print("A spectrogram (Notebook 1) shows us which frequencies are present at each moment.")
```

## 9. Conditions — Making Decisions

Sometimes you need to make decisions based on your data. Python uses `if`, `elif` (else if), and `else` for this:

```python
def classify_singer(song_count):
    """Classify a species by how many songs were recorded."""
    if song_count >= 30:
        return "Prolific singer"
    elif song_count >= 15:
        return "Moderate singer"
    else:
        return "Occasional singer"

# Test it with our data
birds  = ["Bachman's Sparrow", "Northern Cardinal", "Song Sparrow",
          "Swamp Sparrow", "Zebra Finch"]
counts = [19, 9, 15, 10, 45]

for bird, count in zip(birds, counts):
    category = classify_singer(count)
    print(f"{bird}: {count} songs -> {category}")
```
